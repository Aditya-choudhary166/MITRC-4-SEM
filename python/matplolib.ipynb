```python
##bar plot
x=["mp","up","rj","hm"]
y=[1000,2000,3000,4000]
colours=["red","green","blue","yellow"]
plt.bar(x,y,color=colours,width=0.3,
        edgecolor = "black",
        label = x
        )
plt.legend() #square box mai hai jo
plt.grid() #grid line show karega
        
# plt.bar(x,y)  
```

```python
##histogram == distribution of data in bins(interval)
data = np.random.randn(1000)
plt.hist(data,bins=30)
plt.show()
```

```python
##pie chart--> distribution of data in slices
labels=["a","b","c","d"]
sizes=[10,20,30,40]
explodes=(0,0,1,0)
plt.pie(sizes,labels=labels,autopct="%1.2f%%",shadow=True,explode=explodes)
plt.title("pie chart")
plt.savefig("plot.png")
plt.show()
```

```python
##subplots --multiple plots in a single  figure

x=np.linspace(0,10,100)
plt.figure(figsize=(10,5)) #10,5 is a x,y 
# plt.subplot(2,2,1) #2 rows 2 columns 1st subplot
plt.plot(x,np.sin(x))
plt.title("sin(x)")


#cos(x) graph
plt.subplot(2,2,2) #2 rows 2 columns 1st subplot
plt.plot(x,np.cos(x))
plt.title("cos(x)")


plt.subplot(2,2,3) #2 rows 2 columns 1st subplot
plt.plot(x,np.exp(-x))
plt.title("exp(-x)")


plt.subplot(2,2,3) #2 rows 2 columns 1st subplot
plt.plot(x,np.tan(x))
plt.title("tan(x)")



```

```python
plt.tight_layout()
plt.show()
```
